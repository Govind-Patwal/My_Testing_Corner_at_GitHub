```python
import pandas as pd
```

```python
shopping_df = pd.read_csv("resources/shopping_data.csv")
shopping_df
```

```python
shopping_df.columns
```

```python
shopping_df.dtypes
```

```python
# find null values
for column in shopping_df.columns:
    print(f"Column {column} has {shopping_df[column].isnull().sum()} null values")
```

```python
# drop null rows
shopping_df = shopping_df.dropna()
shopping_df
```

```python
# print duplicate entries
print(f"Duplicate Enries = {shopping_df.duplicated().sum()}")
```

```python
# Remove the column CustomerID
shopping_df = shopping_df.drop('CustomerID', axis = 1)
shopping_df
```

```python
# function to change Caard Member to 1 and non-card member to 0
def change_string (member):
  if member == "Yes":
    return 1
  else:
    return 0

shopping_df["Card Member"] = shopping_df["Card Member"].apply(change_string)
shopping_df
```

```python
shopping_df['Annual Income'] = shopping_df['Annual Income']/1000
shopping_df
```

```python
#renaming many columns in one go
shopping_df.columns = ['CardMember', 'Age', 'AnnualIncome', 'SpendingScore']
shopping_df
```

```python
shopping_df.to_csv("resources/clean_shopping_df.csv", index=False)
```
