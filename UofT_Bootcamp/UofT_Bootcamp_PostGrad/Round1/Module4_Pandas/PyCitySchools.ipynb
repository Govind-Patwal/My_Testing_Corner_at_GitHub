```python
# import dependencies
import pandas as pd
```

```python
student_df = pd.read_csv('Resources/clean_students_complete.csv', index_col=0)
student_df
```

```python
school_df = pd.read_csv('Resources/schools_complete.csv')
school_df
```

```python
total_students = len(student_df)
total_students
```

```python
total_schools = len(school_df)
total_schools
```

```python
total_budget = sum(school_df['budget']) # or school_df['Budget'].sum()
total_budget
```

```python
total_math_score = sum(student_df['math_score'])
average_math_score = total_math_score/total_students
average_math_score

# alternatively
# student_df['math_score'].mean()
```

```python
total_reading_score = sum(student_df['reading_score'])
average_reading_score = total_reading_score/total_students
average_reading_score
```

```python
combined_df = student_df.merge(school_df, on ='school_name')
combined_df
```

```python
combined_df.count()
```

```python
student_count = combined_df['student_name'].count()
```

```python
school_count = len(combined_df['school_name'].unique())
school_count
```

```python
# determine the passing grade
```

```python
count_math_pass = len(combined_df.loc[combined_df['math_score'] >= 70])
count_math_pass
```

```python
count_reading_pass = len(combined_df.loc[combined_df['reading_score'] >= 70])
count_reading_pass
```

```python
count_math_and_reading_pass = len(combined_df.loc[(combined_df['math_score'] >= 70) & (combined_df['reading_score'] >= 70)])
count_math_and_reading_pass
```

```python
percentage_math_pass = count_math_pass/total_students * 100
percentage_math_pass
```

```python
percentage_reading_pass = count_reading_pass/total_students * 100
percentage_reading_pass
```

```python
percentage_math_and_reading_pass = count_math_and_reading_pass/total_students * 100
percentage_math_and_reading_pass
```

```python
school_summary_df = pd.DataFrame([{
    'Total Student': f'{total_students:,}',
    'Total_Schools' : f'{school_count:,}',
    'total Budget' :f'${total_budget:,.2f}',
    'Count Math Pass' : f'{count_math_pass:,}',
    'Count Reading Pass': f'{count_reading_pass:,}',
    'Count Reading & Math Pass' : f'{count_math_and_reading_pass:,}',
    'Percentage Math Pass' : f'{percentage_math_pass:.2f}',
    'Percentage Reading Pass' : f'{percentage_reading_pass:.2f}',
    'Percentage Math & Reading Pass' : f'{percentage_math_and_reading_pass:.2f}'
}])

school_summary_df
```
